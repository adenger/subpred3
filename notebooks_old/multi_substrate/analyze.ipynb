```python
from pathlib import Path
import pandas as pd

cols = ["feature", "feature_selection", "substrate",  "dataset", "score"]
records = []

for file in Path("results").glob("*scores.tsv"):
    feature, feature_selection, _ = file.name.split("_")
    df = pd.read_table(file, index_col=0, skiprows=1).drop(["label"])
    df = pd.melt(df.reset_index(), id_vars=["dataset"])
    for row in df.values:
        records.append(
            [feature, feature_selection] + row.tolist()
        )

df = pd.DataFrame.from_records(data=records, columns=cols)
df_piv =df.pivot_table(index=["feature", "feature_selection"], columns=["substrate", "dataset"] )
df_piv = df_piv.loc[["AAC", "PAAC", "PSSM", "Combined"], "score"]



df_piv = df_piv.sort_index(axis=1, ascending=False)[["Amino-acid", "Electron", "Potassium", "Sugar", "mean"]]

display(df_piv)
print(df_piv.to_latex())
```
